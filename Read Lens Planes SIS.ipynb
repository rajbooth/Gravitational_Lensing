```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import healpy as hp
import math
from ipywidgets import *
from IPython.display import display
from astropy.table import Table, Column
import h5py
import os
```

```python
particle = np.dtype([('id',np.uint32),('x',np.float32),('y',np.float32),('z',np.float32),         
                    ('vx',np.float32),('vy',np.float32),('vz',np.float32),
                    ('ra',np.float32),('dec',np.float32),('zred',np.float32)])
```

```python
f = open('/lustre/scratch/astro/rb460/lensplanes/SISHaloLCParticles.dat','rb')
num=np.fromfile(f, dtype=np.int32,count=1)
num[0]
```

```python
particledata = np.fromfile(f,dtype=particle, count = num[0])
```

```python
tab=Table(particledata[::10])
col=tab.columns['id']
col.format = "%4d"
col=tab.columns['x']
col.format = "%4.1f"
col=tab.columns['y']
col.format = "%4.1f"
col=tab.columns['z']
col.format = "%4.1f"
col=tab.columns['ra']
col.format = "%4.2f"
col=tab.columns['dec']
col.format = "%4.2f"
col=tab.columns['zred']
col.format = "%4.2f"
#tab
```

```python
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

x = tab.columns['x']
y = tab.columns['y']
z = tab.columns['z']

ax.scatter(x, y, z, s=1, c='b', marker='o')
ax.view_init(10,60)

#display(fig)
```

```python
#x=x-1200
r=np.sqrt(x*x+y*y+z*z)
```

```python
fig2 = plt.figure(figsize=(8,6))
# the histogram of the data
n, bins, patches = plt.hist(r, 50, facecolor='blue',edgecolor='black')

plt.xlabel('r')
plt.ylabel('Probability')
plt.grid(True)

plt.show()
```
